```python
data = "C:/Users/ML/Desktop/123/122B1F090/DATASET/AirQuality.csv"
df = pd.read_csv(data);
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
for column in df.columns:
    if df[column].dtype in ['float64', 'int64']:  # Check if the column is numeric
        median_value = df[column].median()
        df[column].fillna(median_value, inplace=True)  # Use median for numeric columns

    elif df[column].dtype == 'object':  # Check for categorical columns
        mode_value = df[column].mode()[0]
        df[column].fillna(mode_value, inplace=True)  # Use mode for categorical columns

# Check for remaining missing values
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])  # Only print columns with missing values
```

```python
# Load the dataset with a semicolon delimiter
df = pd.read_csv('C:/Users/ML/Desktop/123/122B1F090/DATASET/AirQuality.csv', delimiter=';')
print(df.columns)
```

```python
# Drop unnecessary columns
df = df.drop(columns=['Unnamed: 15', 'Unnamed: 16'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select columns for boxplot (adjust as necessary)
columns_to_plot = ['CO(GT)', 'PT08.S1(CO)', 'C6H6(GT)', 'T', 'RH', 'AH']

# Create boxplots for each selected column
plt.figure(figsize=(12, 8))
df[columns_to_plot].plot(kind='box', subplots=True, layout=(2, 3), figsize=(15, 10), sharey=False)
plt.suptitle("Boxplots for Selected Air Quality Indicators")
plt.show()
```

```python
# Plot a single boxplot for 'CO(GT)'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='CO(GT)')
plt.title("Boxplot for CO(GT) Levels")
plt.show()
```
